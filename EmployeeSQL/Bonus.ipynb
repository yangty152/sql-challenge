```python
from sqlalchemy import create_engine,text
import pandas as pd
from matplotlib import pyplot as plt
from pw import password
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/employees')
connection = engine.connect()
```

```python
employee = pd.read_sql('''select emp_no, emp_title_id, first_name, last_name, birth_date, sex, hire_date from employee''',connection)
title = pd.read_sql('''select * from title''',connection)
salary = pd.read_sql('''select * from salaries''',connection)
```

```python
# Most of the salaries are between 40K to 50K based on the histogram below.
plt.figure(figsize =(10, 5))
plt.hist(salary['salary'], bins = [40000, 50000,60000,70000,80000,90000,100000])
plt.title("Salary Histogram")
plt.xlabel("Salary Range")
plt.ylabel("Employee Count")
plt.show()
```

```python
# Show the average salary per title
title_employee = pd.merge(employee,title, left_on=['emp_title_id'], right_on=['title_id'], how = 'left')
title_employee_salary=pd.merge(title_employee, salary, on = ['emp_no'], how='left')
avg_salarybytitle=title_employee_salary.groupby('title').mean()
avg_salarybytitle=avg_salarybytitle.reset_index()
plt.subplots(figsize =(10, 5))
plt.bar(avg_salarybytitle['title'], avg_salarybytitle['salary'])
plt.title("Average Salary by Title")
plt.xlabel("Title")
plt.ylabel("Average Salary")
plt.show()
```
